# Tensorflow2.0 小练习

```python
import tensorflow as tf
import numpy as np
```

## 实现softmax函数

```python
def softmax(x):
    ##########
    '''实现softmax函数，只要求对最后一维归一化，=>axis=-1
    不允许用tf自带的softmax函数'''
    ##########
    #tf.nn.softmax是直接调用
    exp_x=np.exp(x)
    #print(exp_x)
    prob_x=exp_x/np.sum(exp_x,axis=-1, keepdims=True) #二维数组中axis=-1代表沿着行方向求和，keepdims=True代表保持原始数组维度
    #print(prob_x)
    return prob_x

test_data = np.random.normal(size=[10, 5])
print("test_data:\n",test_data)
#print(softmax(test_data))#2026.3.15
#print(tf.nn.softmax(test_data, axis=-1).numpy())#2026.3.15
(softmax(test_data)- tf.nn.softmax(test_data, axis=-1).numpy())**2 <0.0001
```

## 实现sigmoid函数

```python
def sigmoid(x):
    ##########
    '''实现sigmoid函数， 不允许用tf自带的sigmoid函数'''
    ##########
    prob_x=1/(1+np.exp(-x))
    return prob_x

test_data = np.random.normal(size=[10, 5])
#print(test_data)
#print(sigmoid(test_data))#2026.3.15
#print(tf.nn.sigmoid(test_data).numpy())#2026.3.15
(sigmoid(test_data) - tf.nn.sigmoid(test_data).numpy())**2 < 0.0001
```

## 实现 softmax 交叉熵loss函数

```python
def softmax_ce(x, label):
    ##########
    '''实现 softmax 交叉熵loss函数， 不允许用tf自带的softmax_cross_entropy函数'''
    ##########
    #print(label)
    #print(x)
    #print(np.log(x))
    loss=np.sum(-label*np.log(x),axis=-1,keepdims=False)
    print(loss,loss.shape,loss.dtype)
    return np.mean(loss)

test_data = np.random.normal(size=[10, 5])#test_data 10个样本，每个样本有5个特征值。
prob = tf.nn.softmax(test_data)#softmax处理过
label = np.zeros_like(test_data)#初始化全零数组
label[np.arange(10), np.random.randint(0, 5, size=10)]=1.#随机标定一个样本标签--10个样本，每个样本属于5个类别之一，并且采用了one-hot编码。
#print(label,label.shape,prob.shape)
#loss=softmax_ce(prob,label)
print(tf.nn.softmax_cross_entropy_with_logits(label, test_data))
((tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(label, test_data))- softmax_ce(prob, label))**2 < 0.0001).numpy()
```

## 实现 sigmoid 交叉熵loss函数

```python
def sigmoid_ce(x, label):
    ##########
    '''实现 softmax 交叉熵loss函数， 不允许用tf自带的softmax_cross_entropy函数'''
    ##########
    loss=-(label*np.log(x)+(1-label)*np.log(1-x))
    print("my sigmoid:\n",loss,loss.shape,loss.dtype)
    return np.mean(loss)

test_data = np.random.normal(size=[10])
#print(test_data)
prob = tf.nn.sigmoid(test_data)
#print(prob)
label = np.random.randint(0, 2, 10).astype(test_data.dtype)
print ("label:\n",label)
print("tf sigmoid:\n",tf.nn.sigmoid_cross_entropy_with_logits(label, test_data))
((tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(label, test_data))- sigmoid_ce(prob, label))**2 < 0.0001).numpy()
```

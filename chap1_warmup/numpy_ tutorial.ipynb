```python
#逐元素乘法--两个数组必须能够通过广播兼容。如果形状不一致，NumPy会尝试将较小的数组“广播”到较大数组的形状。
print(x*y)
print(np.multiply(x,y))
#矩阵乘法--需要满足矩阵乘法对维度匹配的要求。
print(np.dot(x,y))
```

##### 19.利用13题目中的 x,进行求和。提示：输出三种求和 (1)print(np.sum(x)):   (2)print(np.sum(x，axis =0 ));   (3)print(np.sum(x,axis = 1))

```python
#对数组所有元素求和，返回一个标量。
print(np.sum(x))
#将axis=0对应的维度（行）压缩，即对每一列的元素求和。
print(np.sum(x,axis=0))
#将axis=1对应的维度（列）压缩，即对每一行的元素求和。
print(np.sum(x,axis=1))
```

#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
import matplotlib.pyplot as plt

#生成x值（从0到100，步长0.1）
x=np.arange(0, 100, 0.1)

#计算对应的y值
y=x*x#x**2

#绘制图形
plt.plot(x, y)

#添加标题和坐标轴标签
plt.title("y = x*x")
plt.xlabel("x")
plt.ylabel("y")

# 显示图形
plt.show()
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
x=np.arange(0, 3*np.pi,0.1)

#计算对应的y值
y = np.sin(x)

#绘制图形
plt.plot(x, y)
plt.title("y = sinx")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

#计算对应的y值
y = np.cos(x)

#绘制图形
plt.plot(x, y)
plt.title("y = cosx")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
